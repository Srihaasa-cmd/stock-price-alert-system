```python
import requests    # used to call the API and get data from internet
import pandas as pd    # used to organise data into tables
import matplotlib.pyplot as plt    # used to draw charts
import sqlite3    # used to store data in a database on your laptop
from datetime import datetime    # used to get current date and time
```

```python
# Your Alpha Vantage API key
API_KEY = "D69MMB57GGIV097S"  # replace this with your actual key

# The stocks you want to track
# .BSE means Bombay Stock Exchange (Indian stocks)
STOCKS = {
    'Reliance'  : 'RELIANCE.BSE',
    'TCS'       : 'TCS.BSE',
    'Infosys'   : 'INFY.BSE',
    'HDFC Bank' : 'HDFCBANK.BSE',
    'Wipro'     : 'WIPRO.BSE'
}

# Your alert prices
# above = send alert if stock goes HIGHER than this
# below = send alert if stock goes LOWER than this
ALERTS = {
    'Reliance'  : {'above': 3000, 'below': 2500},
    'TCS'       : {'above': 4500, 'below': 3800},
    'Infosys'   : {'above': 1800, 'below': 1500},
    'HDFC Bank' : {'above': 1800, 'below': 1500},
    'Wipro'     : {'above': 600,  'below': 450},
}
```

```python
# This function fetches stock data for ONE stock
def fetch_stock(stock_name, ticker):
    
    # Step 1 — Build the request
    # Think of this like filling a form before calling the waiter
    params = {
        'function'   : 'TIME_SERIES_DAILY',  # give me daily prices
        'symbol'     : ticker,               # which stock
        'apikey'     : API_KEY,              # prove it's me
        'outputsize' : 'compact'             # last 100 days only
    }
    
    # Step 2 — Send the request to Alpha Vantage
    # requests.get = calling the waiter
    url = "https://www.alphavantage.co/query"
    response = requests.get(url, params=params)
    
    # Step 3 — Convert response to Python dictionary
    data = response.json()
    
    # Step 4 — Check if data came back correctly
    if 'Time Series (Daily)' not in data:
        print(f"Error fetching {stock_name}")
        print(f"Message: {data}")
        return None
    
    # Step 5 — Convert to pandas dataframe
    time_series = data['Time Series (Daily)']
    df = pd.DataFrame.from_dict(time_series, orient='index')
    
    # Step 6 — Clean up column names
    df.columns = ['open', 'high', 'low', 'close', 'volume']
    df = df.astype(float)   # convert text numbers to real numbers
    df.index = pd.to_datetime(df.index)   # convert text dates to real dates
    df = df.sort_index()    # sort oldest to newest
    
    print(f"{stock_name} fetched! Latest price: ₹{df['close'].iloc[-1]:.2f}")
    return df

# Now fetch ALL stocks using the function above
all_data = {}
for name, ticker in STOCKS.items():
    df = fetch_stock(name, ticker)
    if df is not None:
        all_data[name] = df

print(f"\nSuccessfully fetched {len(all_data)} stocks!")
```

```python
conn = sqlite3.connect('stocks.db')
cursor=conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS stock_prices (
        stock_name  TEXT,
        date        TEXT,
        open_price  REAL,
        high_price  REAL,
        low_price   REAL,
        close_price REAL,
        volume      REAL,
        saved_at    TEXT
    )
''')
for stock_name, df in all_data.items():
    
    # Loop through every row (every day) in the dataframe
    for date, row in df.iterrows():
        cursor.execute('''
            INSERT INTO stock_prices
            (stock_name, date, open_price, high_price, low_price, close_price, volume, saved_at)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?)
        ''', (
            stock_name,          # eg. Reliance
            str(date),           # eg. 2024-01-01
            row['open'],         # opening price that day
            row['high'],         # highest price that day
            row['low'],          # lowest price that day
            row['close'],        # closing price that day
            row['volume'],       # how many shares traded
            datetime.now().strftime('%Y-%m-%d %H:%M:%S')  # when we saved this
        ))

# Step 5 — Commit saves all changes permanently to the database
# Like pressing Ctrl+S in Excel
conn.commit()

# Step 6 — Close the connection
# Like closing the Excel file after saving
conn.close()

print("Data saved to database!")
```

```python
def analyse_stock(stock_name, df):
    
    # Step 1 — Get the latest closing price
    # iloc[-1] means last row = most recent day
    latest_price = df['close'].iloc[-1]
    
    # Step 2 — Calculate moving averages
    # Moving average = average of last N days closing prices
    # It smooths out daily ups and downs to show the real trend
    ma7  = df['close'].rolling(window=7).mean().iloc[-1]   # average of last 7 days
    ma21 = df['close'].rolling(window=21).mean().iloc[-1]  # average of last 21 days
    
    # Step 3 — Calculate weekly price change
    # iloc[-5] = price 5 trading days ago = roughly 1 week ago
    week_ago_price  = df['close'].iloc[-5]
    weekly_change   = ((latest_price - week_ago_price) / week_ago_price * 100).round(2)
    
    # Step 4 — Generate buy/sell signal
    # If short term average (7day) is ABOVE long term average (21day)
    # = price is trending UP = BULLISH = good time to buy
    # If short term is BELOW long term
    # = price is trending DOWN = BEARISH = careful
    if ma7 > ma21:
        signal = 'BULLISH'
        advice = 'Short term trend going UP'
    else:
        signal = 'BEARISH'
        advice = 'Short term trend going DOWN'
    
    # Step 5 — Check if alert should trigger
    alert = ALERTS[stock_name]
    alert_triggered = None
    
    if latest_price > alert['above']:
        alert_triggered = f"{stock_name} is ABOVE ₹{alert['above']} — currently ₹{latest_price:.2f}"
    elif latest_price < alert['below']:
        alert_triggered = f"{stock_name} is BELOW ₹{alert['below']} — currently ₹{latest_price:.2f}"
    
    # Step 6 — Print the analysis
    print(f"\n{'='*45}")
    print(f"  {stock_name}")
    print(f"{'='*45}")
    print(f"  Latest Price   : ₹{latest_price:.2f}")
    print(f"  Weekly Change  : {weekly_change}%")
    print(f"  7-day MA       : ₹{ma7:.2f}")
    print(f"  21-day MA      : ₹{ma21:.2f}")
    print(f"  Signal         : {signal} — {advice}")
    
    if alert_triggered:
        print(f"  ALERT          : {alert_triggered}")
    else:
        print(f"  Alert          : Within normal range")
    
    # Step 7 — Return results as dictionary
    return {
        'stock'           : stock_name,
        'latest_price'    : round(latest_price, 2),
        'weekly_change'   : weekly_change,
        'ma7'             : round(ma7, 2),
        'ma21'            : round(ma21, 2),
        'signal'          : signal,
        'alert_triggered' : alert_triggered
    }

# Run analysis on ALL stocks
results = []
for name, df in all_data.items():
    result = analyse_stock(name, df)
    results.append(result)

# Convert results to a clean dataframe
results_df = pd.DataFrame(results)
print("\n")
print("=== SUMMARY TABLE ===")
print(results_df[['stock', 'latest_price', 'weekly_change', 'signal']])
```

```python
# Create one chart per stock
fig, axes = plt.subplots(1, len(all_data), figsize=(14, 5))

# If only 1 stock axes won't be a list — fix that
if len(all_data) == 1:
    axes = [axes]

colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']

for i, (name, df) in enumerate(all_data.items()):
    ax = axes[i]
    
    # Plot closing price
    ax.plot(df.index, df['close'],
            color=colors[i], linewidth=1.5, label='Price')
    
    # Plot 7 day moving average
    ax.plot(df.index, df['close'].rolling(7).mean(),
            color='orange', linewidth=1,
            linestyle='--', label='7-day MA')
    
    # Plot 21 day moving average
    ax.plot(df.index, df['close'].rolling(21).mean(),
            color='red', linewidth=1,
            linestyle='--', label='21-day MA')
    
    # Draw alert lines
    ax.axhline(y=ALERTS[name]['above'],
               color='green', linewidth=0.8,
               linestyle=':', label='Upper alert')
    ax.axhline(y=ALERTS[name]['below'],
               color='red', linewidth=0.8,
               linestyle=':', label='Lower alert')
    
    # Labels
    ax.set_title(f'{name}', fontsize=12, fontweight='bold')
    ax.set_ylabel('Price (₹)')
    ax.legend(fontsize=7)
    ax.grid(axis='y', linestyle='--', alpha=0.4)
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Stock Price Monitor — NSE India', fontsize=14)
plt.tight_layout()
plt.savefig('stock_monitor.png')
plt.show()
```
